```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
filepath= "../input/one-piece-arcs/OnePieceArcs.csv"
op_data = pd.read_csv(filepath, index_col="Arc")
```

<h2>Finding the Arc with the most Anime Episode</h2>

```python
arc_most_episode = op_data[op_data['TotalEpisodes'] == op_data.TotalEpisodes.max()]
arc_most_episode
```

```python
#This is odd. As a One Piece fan, I know that Wano arc is the longest. Let's check Wano arc.
op_data.tail()
```

```python
#Wano Arc was divided into three. We need to combine them.
wano_combined = op_data.loc[["Wano Country Arc: Act 3", "Wano Country Arc: Act 2", "Wano Country Arc: Act 1"]]

wanoTotalChapters = sum(wano_combined['TotalChapters'])
wanoTotalPages = sum(wano_combined['TotalPages'])
wanoMangaPercent = "{:.1f}".format(wanoTotalChapters/sum(op_data['TotalChapters'])*100)+"%"

wanoTotalEpisode = sum(wano_combined['TotalEpisodes'])
wanoTotalMinutes = sum(wano_combined['TotalMinutes(avg 24)'])
wanoAnimePercent = "{:.1f}".format(wanoTotalEpisode/sum(op_data['TotalEpisodes'])*100)+"%"
```

```python
##Adding Wano to Dataframe
op_data.loc[len(op_data.index)] = [909,wanoTotalChapters,wanoTotalPages,wanoMangaPercent,890,wanoTotalEpisode,wanoTotalMinutes,wanoAnimePercent]
op_data.tail()
```

```python
## Let's Rename it back to Wano
op_data = op_data.rename(index={51: "Wano"})

##Removing other Wano arcs
op_data =op_data.drop(index=['Wano Country Arc: Act 1', 'Wano Country Arc: Act 2', 'Wano Country Arc: Act 3'])
op_data.tail()
```

```python
##Let's find the arc with the most episode
arc_most_episode = op_data[op_data['TotalEpisodes'] == op_data.TotalEpisodes.max()]
print(f"The arc with one piece with the most episode is {arc_most_episode.index.values} with {op_data.TotalEpisodes.max()} episodes")
```

## Removing % symbol from manga and anime

```python
def filter_percent(data):
    return data[:-1]
```

```python
op_data['Manga%'] = op_data['Manga%'].apply(filter_percent).astype(float)
op_data['Anime%'] = op_data['Anime%'].apply(filter_percent).astype(float)
```

```python
##Checking Code
op_data.head()
```

## Total Episodes per arc %

```python
plt.figure(figsize=(14,10))
plt.title("Total Episdoes per Arc")
sns.barplot(x=op_data["Anime%"],y=op_data.index)
```

### One Piece has a bad pacing problem. Let's compare the Manga chapters to Anime episodes per arc

```python
op_data['TotalChapters'] = op_data['TotalChapters'].astype(int)
op_data['TotalEpisodes'] = op_data['TotalEpisodes'].astype(int)
op_data["Anime to Manga Ratio"] = op_data['TotalEpisodes']/op_data['TotalChapters']
op_data.head()
```

```python
plt.figure(figsize=(14,12))
plt.title("Anime to Manga Ratio per Arc")
sns.barplot(x=op_data["Anime to Manga Ratio"], y=op_data.index)
```
